```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数


def v(x):
    numerator1 = 1333.2
    denominator1 = 15000 * 101325
    factor1 = numerator1 / denominator1

    exponent = (-1.45*1.6e-19 / (1.38e-23)) * x
    exp_value = np.exp(exponent)

    numerator2 = 0.4 * 5e6 * exp_value
    denominator2 = 0.4 + 5e6 * exp_value

    factor2 = numerator2 / denominator2

    return factor1 * factor2


# 定义 x 的范围
x = np.linspace(1/1200, 1/500, 1000)

# 计算 y 值
y = v(x)

# 绘图
plt.plot(x, y)
plt.xlabel('1/T')
plt.ylabel('v(T)')
plt.title('Plot of v(T)')
plt.grid(True)
plt.show()
```
